```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower() #replace all letter into lowercase
df.columns = df.columns.str.replace(' ', '_')  #replace spaces into underscore
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
# qcut= splits ages into 4 equal size group based on data distribution & assign the label we defined
```

```python
df.columns
```

```python
df[['age', 'age_group']].head(7)
```

```python
# create a column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,   # fixed spelling
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = (
    df['frequency_of_purchases']
    .str.strip()
    .map(frequency_mapping)
    .fillna(0)          # removes NaN
    .astype(int)
)
#map = to replace the text in the column with cooresponding no.of days
```
